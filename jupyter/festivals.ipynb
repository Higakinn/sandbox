```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = 'http://www.saijomatsuri.jp/yatai/'
site = requests.get(url)
data = BeautifulSoup(site.text, 'html.parser')
div = data.find("div", {"class":"entry-ya-content"})
bs_yatai = data.findAll("table", {"class":"yatalist"})
dict_yatai = {}
for i, bs_area in enumerate(div.findAll("p")):
  bs_yatai_img = bs_yatai[i].findAll("img")
  yatai_list = [{"name":y.get('title'),"img":y.get('src')} for y in bs_yatai_img]
  dict_yatai[bs_area.get_text()] = yatai_list
```
